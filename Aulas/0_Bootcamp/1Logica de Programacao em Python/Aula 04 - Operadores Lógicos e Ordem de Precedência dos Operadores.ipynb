#### **Operadores Lógicos**

Operador | Nome
-------- | --------
not      | Negação
and      | Conjunção
or       | Disjunção

```python
# not 

x = 1

print(not x == 1)
```

```python
# and 

x = 1

print(x > 1 and x < 100)
```

```python
# or

x = 1

print(x > 200 or x > 100)
```

#### **Ordem de Precedência dos Operadores**
* Parênteses
* Exponenciação
* Multiplicação, divisão, módulo e divisão inteira
* Adição e subtração, que possuem a mesma precedência

(3 - 1) + 4 ** 2 + 5 * 9 % 5

2 + 4 ** 2 + 5 * 9 % 5

2 + 16 + 45 % 5
