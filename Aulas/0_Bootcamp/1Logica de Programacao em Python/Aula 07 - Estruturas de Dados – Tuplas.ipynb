# **Aula 07 - Estruturas de Dados – Tuplas**

<br>
<br>

___
Tupla é um tipo de estrutura de dados utilizada em Python que funciona de modo semelhante a uma lista, entretanto, com a característica principal de ser imutável.

Uma tupla consiste em uma sequência de valores separados por vírgulas.

**Criando uma Tupla**

```python
tupla = ()
#
print(type(tupla))
```

```python
tupla = tuple()

print(type(tupla))
```

```python
tupla = ('Sirius', 'Bootcamp')

print(tupla)
```

```python
tupla = 'Sirius', 'Bootcamp', 1

print(type(tupla))
```

**Acessando Elementos da Tupla**

```python
tupla = ('João', 18)
```

```python
print(tupla[0], tupla[1])
```

```python
tupla[0] = 'Pedro'
```

```python
numeros = (1,4,5,5,6,7,4,9,2)
```

```python
numeros.count(4)
```

```python
numeros.index(2)
```

```python
tuplas = ([1,2,3], ['a', 'b', 'c'])

print(tuplas)
```

```python
lista = [('João', 18), ('Carlos', 20)]

print(lista)
```
