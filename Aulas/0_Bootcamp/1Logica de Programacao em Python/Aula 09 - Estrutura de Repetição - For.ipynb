Um **loop for** é uma instrução para iterar sobre uma sequência (lista, dicionário, entre outros), permitindo que o código seja executado várias vezes.

for.png


```python
mensagem = 'Bem-vindo a Sirius'
#
```

```python
print(mensagem, mensagem, mensagem, mensagem, mensagem)
```

```python
print(mensagem)
print(mensagem)
print(mensagem)
print(mensagem)
print(mensagem)
```

```python
for i in range(5):
  print(mensagem)
```

```python
range(5)
```

```python
for i in range(5):
  print(i)
```

**Exercício:** Escreva um programa que solicita ao usuário 5 números. Após o usuário digitar os 5 números, o programa deve exibir a soma dos números:

```python
numero_1 = int(input('Digite o 1 número: '))
numero_2 = int(input('Digite o 2 número: '))
numero_3 = int(input('Digite o 3 número: '))
numero_4 = int(input('Digite o 4 número: '))
numero_5 = int(input('Digite o 5 número: '))
```

```python
soma = numero_1 + numero_2 + numero_3 + numero_4 + numero_5

print(f'A soma é igual a {soma}')
```

```python
soma = 0
for i in range(5):
  numero = int(input('Digite um número: '))

  soma = soma + numero

print(f'A soma é igual a {soma}')
```

```python
lista = [1,2,3,4,5]

for i in lista:
  print(i)
```

```python
notas = {'Matemática':10, 'Português': 8, 'Algoritmo': 8}
```

```python
for i in notas.items():
  print(i)
```

```python
for chave, valor in notas.items():
  print(f'Chave = {chave} e Valor = {valor}')
```
