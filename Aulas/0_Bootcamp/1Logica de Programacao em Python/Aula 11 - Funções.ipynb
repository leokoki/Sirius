```python
def soma():
  somar_numeros = 10 + 2

  print(f'A soma é igual a {somar_numeros}')
```

```python
soma()
```

```python
def soma_II(a, b):

  somar_numeros = a + b

  print(f'A soma é igual a {somar_numeros}')
```

```python
soma_II(2, 8)
```

```python
soma_II(10, 20)
```

```python
soma_II(100, 200)
```

```python
def soma_III(a, b):

  soma = a + b

  return soma
```

```python
somar = soma_III(30, 40)
```

```python
print(f'A soma é igual a {somar}')
```

```python
def soma_IIII(a, b):

  soma = a + b

  media = (a + b) / 2

  return soma, media
```

```python
somar = soma_IIII(100, 20)
```

```python
print(somar)
```

```python
soma, media = soma_IIII(100, 20)
```

```python
print(soma)
```

```python
print(media)
```

```python
def entrada_dados():
  numero_1 = int(input('Digite um número: '))
  numero_2 = int(input('Digite outro número: '))

  soma = soma_III(numero_1, numero_2)

  print(f'A soma é igual a {soma}')
```

```python
entrada_dados()
```
