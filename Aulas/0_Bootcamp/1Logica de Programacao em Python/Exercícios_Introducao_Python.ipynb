## Exercício 1 (Aquecimento): Vamos criar uma calculadora (simples) com as operações soma, subtração, multiplicação, divisão.

Dica: crie funções para tarefas específicas como, por exemplo, uma função `soma(a,b)`, uma função `subtracao(a,b)`, ... e depois use estas funções para criar a função `calculadora()`.

Lembre-se de que não se pode dividir por zero.

```python
# seu código
def soma(a,b):
  return a+b

def subtracao(a,b):
  return a-b

def multiplicacao(a,b):
  return a*b

def divisao(a,b):
  if b == 0:
    return("Não se pode dividir por zero!")
  else:
    return a/b
```

```python
# seu código

def calculadora():
  continua=True
  sinal={
      1:'+',
      2:'-',
      3:'x',
      4:'÷'
  }
  while continua:
    opcao=0
    while opcao not in [1,2,3,4,5]:
      print('-'*100)
      print('-'*100)
      opcao = int(input('''Escolha o número da operação a ser realizada:
        \n\t1) soma
        \n\t2) subtração
        \n\t3) multiplicação
        \n\t4) divisão
        \n\t5) encerrar calculadora\n
        '''))
      if opcao not in [1,2,3,4,5]:
        print('Opção inválida, tente novamente')
    if opcao ==5:
      break
    a = int(input('Valor de A\n'))
    b = int(input('Valor de B\n'))
    if 1==opcao: 
      resultado = soma(a,b)
    elif 2==opcao: 
      resultado = subtracao(a,b)
    elif 3==opcao: 
      resultado = multiplicacao(a,b)
    else: 
      resultado = divisao(a,b)
    if opcao in [1,2,3,4]:
      print(f'O resultado de {a} {sinal[opcao]} {b} = {resultado}')
    print('-'*100)
    print('-'*100)
  print('Calculadora Encerrada') 
```

```python
calculadora()
```
