12 - Não é obrigatório, mas não menos importante, você conseguiria fazer um gráfico de barras com a contagem que fez acima?

Você pode consultar aqui a documentação do pandas e gráficos de barras:
    
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html

```python
import plotly.express as px
```

```python
fig = px.bar(df.Estado.value_counts().reset_index().rename(columns={'index':'Estado','Estado':'Quantidade'})
              , x='Estado',y='Quantidade',
             text_auto='Quantidade'
             )
fig.update_layout(title_text='Pessoas por Estado', title_x=0.5)
fig.show()
```
